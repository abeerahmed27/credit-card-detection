# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
dataset['std_Amount'] = sc.fit_transform(dataset['Amount'].values.reshape (-1,1))
dataset = dataset.drop("Amount", axis=1) #removing Amount
```

```python
sns.countplot(x="Class", data=dataset)
```

```python
import imblearn #To undersample, we can use the package imblearn with RandomUnderSampler function 
from imblearn.under_sampling import RandomUnderSampler 

undersample = RandomUnderSampler(sampling_strategy=0.5)
```

```python
cols = dataset.columns.tolist()
cols = [c for c in cols if c not in ["Class"]]
target = "Class"
```

```python
X = dataset[cols]
Y = dataset[target]

# I would include a PCA convertion of the X variables; as there are some high correlation between some of the varibles
# In that way we can avoid colineality:

from sklearn.decomposition import PCA
pca=PCA(n_components=15, svd_solver='full')
pca = pca.fit_transform(X)

J = pd.DataFrame(pca)
```

```python
#undersample
X_under, Y_under = undersample.fit_resample(X, Y)
X_pca, Y_pca = undersample.fit_resample(J, Y)
```

```python
from pandas import DataFrame
test = pd.DataFrame(Y_under, columns = ['Class'])
```

# Visualizing undersampling results

```python
fig, axs = plt.subplots(ncols=2, figsize=(13,4.5))
sns.countplot(x="Class", data=dataset, ax=axs[0])
sns.countplot(x="Class", data=test, ax=axs[1])

fig.suptitle("Class repartition before and after undersampling")
a1=fig.axes[0]
a1.set_title("Before")
a2=fig.axes[1]
a2.set_title("After")
```

# Training the model

```python
#train the model
from sklearn.linear_model import LogisticRegression # performing logistic regression 
classifier = LogisticRegression(random_state = 2)
classifier.fit(X_trainP, Y_trainP)

```

```python
#predictions
Y_pred = classifier.predict(X_testP)
```

# Elbow method

```python
x = dataset.iloc[:,1: ].values

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 50)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

# Visulaising the graph to observe the elbow

```python
plt.figure(figsize=(12,6))
plt.grid
plt.plot(range(1, 11), wcss, color = 'magenta', marker = "8")
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```
